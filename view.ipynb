```python
from pathlib import Path
import pandas as pd

# 1. 파일 경로 설정
xpt_dir = Path("D:/3 교육/부트캠프/0 wassup ai/Project 1")

# 2. .xpt 파일 전체 읽기 (소문자 확장자)
xpt_files = sorted(list(xpt_dir.glob("*.xpt")))

print(f"총 {len(xpt_files)}개 파일 발견됨")

# 3. 병합
if not xpt_files:
    print("❌ .xpt 파일이 없습니다. 경로와 확장자 다시 확인하세요.")
else:
    df_total = pd.read_sas(xpt_files[0], format='xport')
    print(f"Loaded: {xpt_files[0].name}, shape: {df_total.shape}")

    for xpt_file in xpt_files[1:]:
        df = pd.read_sas(xpt_file, format='xport')
        print(f"Merging: {xpt_file.name}, shape: {df.shape}")
        df_total = pd.merge(df_total, df, on="SEQN", how="left")

    # 저장
    df_total.to_csv(xpt_dir / "nhanes_merged.csv", index=False)
    print("✅ 병합 완료. 최종 shape:", df_total.shape)
```

```python
df_c = df_dpq.copy()
df_c = df_c.replace([7, 9], np.nan)
df_c
```

```python
# 2. SEQN 분리
seqn = df_c['SEQN']
dpq_only = df_c.drop(columns='SEQN')

# 3. 결측치 2개 이하 필터링
dpq_valid = dpq_only[dpq_only.isnull().sum(axis=1) <= 2].copy()
seqn_valid = seqn[dpq_valid.index]

# 4. 평균 대체 (행별)
dpq_filled = dpq_valid.apply(lambda row: row.fillna(row.mean()), axis=1)

# 5. SEQN 다시 결합
df_dpq2 = pd.concat([seqn_valid.reset_index(drop=True), dpq_filled.reset_index(drop=True)], axis=1)

# 정확히 DPQ010 ~ DPQ090만 포함 (총 9개 문항)
dpq_cols = [f'DPQ0{i:02d}' for i in range(10, 100, 10)]

# PHQ-9 총점 계산
df_dpq2['PHQ9_SUM'] = df_dpq2[dpq_cols].sum(axis=1)
```

```python
def interpret_phq9(score):
    if pd.isnull(score):
        return np.nan
    elif score <= 4:
        return 'None-minimal'
    elif score <= 9:
        return 'Mild'
    elif score <= 14:
        return 'Moderate'
    elif score <= 19:
        return 'Moderately severe'
    else:
        return 'Severe'

df_dpq2['PHQ9_LEVEL'] = df_dpq2['PHQ9_SUM'].apply(interpret_phq9)
```

```python
valid_seqn = df_dpq2['SEQN']
```

```python
# 나머지 DF들도 유효한 SEQN만 남기기
dfs_to_merge = [df_demo, df_slq, df_paq, df_smq, df_alq]
dfs_filtered = [df[df['SEQN'].isin(valid_seqn)] for df in dfs_to_merge]

# 기준 df도 일치하는 SEQN만 필터링 (혹시라도 순서 달라질 수 있으니)
df_main = df_dpq2[df_dpq2['SEQN'].isin(valid_seqn)].copy()
```

```python
from functools import reduce

# 병합할 목록: df_main + 나머지 필터된 dfs
merged_list = [df_main] + dfs_filtered

df_total = reduce(lambda left, right: pd.merge(left, right, on='SEQN', how='inner'), merged_list)

print("최종 병합 완료 ✅")
print("데이터 크기:", df_total.shape)
```
