```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 불러오기
df = pd.read_csv("C:/Users/USER/Downloads/archive (5)/depression_data.csv")

# 시각화 스타일 설정
sns.set(style="whitegrid")

# 1. 정신질환 병력 분포 (Barplot)
plt.figure(figsize=(6, 4))
sns.countplot(x="History of Mental Illness", data=df)
plt.title("Mental Illness History Distribution")
plt.ylabel("Count")
plt.xlabel("History of Mental Illness")
plt.tight_layout()
plt.show()

# 2. 연령 분포 (Histogram)
plt.figure(figsize=(8, 5))
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.tight_layout()
plt.show()

# 3. 소득 분포 (Histogram)
plt.figure(figsize=(8, 5))
sns.histplot(df["Income"], bins=50, kde=True)
plt.title("Income Distribution")
plt.xlabel("Income")
plt.tight_layout()
plt.show()

# 4. 소득 vs 정신질환 (Boxplot)
plt.figure(figsize=(8, 5))
sns.boxplot(x="History of Mental Illness", y="Income", data=df)
plt.title("Income by Mental Illness History")
plt.tight_layout()
plt.show()

# 5. 신체활동 수준 vs 정신질환 병력 (Barplot)
plt.figure(figsize=(8, 5))
sns.countplot(x="Physical Activity Level", hue="History of Mental Illness", data=df)
plt.title("Physical Activity Level by Mental Illness History")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```
